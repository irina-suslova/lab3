```python
import pandas as pd

path = "./testsLab3/"
```

```python
dataframeGeneration = pd.DataFrame()
dataframeGeneration[["graph"]] = pd.read_csv(path + "generationGraph.txt", header = None, sep = '\n')
dataframeGeneration[["digraph"]] = pd.read_csv(path + "generationDigraph.txt", header = None, sep = '\n')
dataframeGeneration[["n"]] = dataframeGeneration.index * 100 + 10
dataframeGeneration = dataframeGeneration.set_index('n', drop = True)
dataframeGeneration.plot(title = "Generation")
```
